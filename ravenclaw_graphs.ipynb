```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
executionDatabaseCSVPath = "Resources/DPIC Execution Database - U.S. Executions.csv"
executionDF = pd.read_csv(executionDatabaseCSVPath)
```

```python
executionDF.head(15)
```

```python
executionDF.info()
```

```python
executionDF['State'].value_counts()
```

```python
# Create a dataframe with only State data
stateDF = executionDF['State'].value_counts()

# Initialize the figure
f, ax = plt.subplots(figsize=(20,8))
sns.set_theme(style="whitegrid", palette="pastel")

# Plot the total number of executions per state
sns.barplot(x=stateDF.values, y=stateDF.index, data=executionDF, orient="h")

# Add a titles
ax.set_title('Number of Executions Per State')
ax.set_xlabel('Execution Count')

# Export and display the figure
plt.tight_layout()
plt.savefig('./Resources/executions_per_state_hbarplot.png', bbox_inches='tight')

plt.show()
```

```python
# Import the Texas Deathrow Records
executionDatabaseCSVPath2 = "Resources/tx_deathrow_fulll.csv"
texasDeathrowDF = pd.read_csv(executionDatabaseCSVPath2)

# Convert any inf values to NaN
texasDeathrowDF.replace([np.inf, -np.inf], np.nan, inplace=True)

# Display the dataframe
texasDeathrowDF.head()
```

```python
texasDeathrowDF.info()
```

```python
# Create a histogram using Seaborn to illustrate Highest Education Levels
#             among deathrow inmates sorted by race
#      1 is completed First Grade - 20 is a Professional Doctorate Degree
sns.set_theme(style= 'whitegrid')
ax1 = sns.displot(texasDeathrowDF, x= 'Highest Education Level', 
            col = 'Race', multiple = 'stack', col_wrap = 3,
            hue = 'Race', palette = 'pastel'
)

# Export and display the figure
plt.savefig('./Resources/race_education_level_histograms.png', bbox_inches='tight')

plt.show()
```

```python
raceCount = texasDeathrowDF['Race'].value_counts()
raceCount
```

```python
# Create a violin plot of Age at Execution by Race
sns.set_style('darkgrid', {"axes.facecolor": ".9"})
plt.figure(figsize=(15, 8))
sns.violinplot(x='Race', y='Age at Execution', data=texasDeathrowDF, 
               bw_adjust=.5, cut=1,linewidth = 1.5, palette='pastel')
plt.title('Age at Execution by Race')
plt.xlabel('Race')
plt.ylabel('Age at Execution')

# Export and Display the figure
plt.tight_layout()
plt.savefig('./Resources/race_execution_age_violin.png', bbox_inches='tight')
plt.show()
```

```python
# Convert 'Execution Date' to datetime format
texasDeathrowDF['Execution Date'] = pd.to_datetime(texasDeathrowDF['Execution Date'], format='%m/%d/%Y')

# Group by year and count the number of executions per year
executions_per_year = texasDeathrowDF['Execution Date'].dt.year.value_counts().sort_index()

# Create a line plot of the number of executions per year
plt.figure(figsize=(18, 5))
sns.lineplot(x=executions_per_year.index, y=executions_per_year.values)
sns.set_context('notebook')
plt.title('Number of Executions Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Executions')

# Export and display the figure
plt.tight_layout()
plt.savefig('./Resources/executions_over_time_lineplot.png', bbox_inches='tight')
plt.show()
```

```python
# Scatter plot of Age at Execution over Time
plt.figure(figsize=(12, 3.5))
sns.set_theme(style='whitegrid')
ax = sns.scatterplot(x='Execution Date', y='Age at Execution', hue='Race', data=texasDeathrowDF, 
                style = 'Race', palette='spring', legend='full')
sns.move_legend(ax, 'upper left', bbox_to_anchor=(1, 1), frameon=False)
plt.title('Age at Execution Over Time by Race')
plt.xlabel('Execution Date')
plt.ylabel('Age at Execution')

# Export and display the figure
plt.tight_layout()
plt.savefig('./Resources/race_age_over_time_scatter.png', bbox_inches='tight')
plt.show()
```
